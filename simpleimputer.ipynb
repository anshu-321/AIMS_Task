```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"Col1":[12, None, 54, 8, None ,2, 72, 23, None, 1],
                   "Col2":[None, 24, 54, 3, None,None, 2, 34, 3, None],
                   "Col3":[25, None, 9 , 13, 23 ,None, 3, 45, None, 8],
                   "Col4":[None, 21, 55, None, 5, 14, 3, None, 13, 8]})

print(df)
```

```python
#simple imputer mean 
def imputer_mean(df):
    colLst=df.columns.tolist()
    for i in colLst:
        df[i].fillna(df[i].mean(),inplace=True)
    return df
```

```python
def imputer_const(df):
    colLst=df.columns.tolist()
    for i in colLst:
        df[i].fillna(float(7.2),inplace=True)
    return df
```

```python
def imputer_median(df):
    colLst=df.columns.tolist()
    for i in colLst:
        df[i].fillna(df[i].median(),inplace=True)
    return df    
```

```python
def SimpleImputer(df,typeof):
    if(typeof=="mean"):
        df = imputer_mean(df)
    elif(typeof=="median"):
        df= imputer_median(df)
    else:
        #replacing some constant value
        df=imputer_const(df)
    return df
```

```python
df=SimpleImputer(df,"mean")
print(df)
```
