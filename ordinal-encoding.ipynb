```python
import numpy as np
import pandas as pd
```

```python
data=({'Brand':["Tata","VW","VW","Honda","Tata","Toyota","Toyota"],
     'Price':[12,14,11,15,21,15,13]})
df=pd.DataFrame(data)
print(df)
```

```python
def ordinal_encode(df):
    df['Brand']=df['Brand'].map({'Tata':0,'VW':1,'Honda':2, 'Toyota':3})
    return df
df=ordinal_encode(df)
print(df)
```
