```python
import pandas as pd
import numpy as np
```

```python
data=({'Brand':["Tata","VW","VW","Honda","Tata","Toyota","Toyota"],
     'Price':[12,14,11,15,21,15,13]})
df=pd.DataFrame(data)
print(df)

def OH_encode(df):
    df_new = pd.get_dummies(df, columns=df.select_dtypes(include='object').columns)
    return df_new
```

```python
df_encoded=OH_encode(df)
print(df_encoded)
```
